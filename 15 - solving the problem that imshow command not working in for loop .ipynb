# when using imshow() command inside for loop the images will not show up ! only the last image will show
# so to solve this issue a special command is used after the imshow() which is show() 
# and that required to import from matplotlib.pyplot import show, imshow 

```python
import cv2
import numpy as np
import glob 
import matplotlib.pyplot as plt 
from matplotlib.pyplot import show, imshow 
```

```python
path =  glob.glob(r'C://Users//MAN//usama//*.jpg')

# to load all the images in the folder 
```

```python
d=1                                # will used as name for saving images
for image in path:
    i = cv2.imread(image)          # reading the image
    i = cv2.cvtColor(i , cv2.COLOR_BGR2GRAY)      # converting the color image to gray scale 
    cv2.imwrite(f'C://Users//MAN//usama1//{d}.jpg',i)  # saving the image 
    plt.imshow( i , cmap = 'gray')  # just to show the result
    show() # without this command the imshow command will show only the last image 
    d += 1             
```
