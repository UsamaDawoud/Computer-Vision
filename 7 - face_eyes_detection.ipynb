```python
import numpy as np 
import cv2 

face_cascade = cv2.CascadeClassifier("C:/Users/MAN/a/deep learning/DATA/haarcascades/haarcascade_frontalface_default.xml") 
eye_cascade = cv2.CascadeClassifier("C:/Users/MAN/a/deep learning/DATA/haarcascades/haarcascade_eye.xml")

cap = cv2.VideoCapture(0)

while True:
    ret , rec = cap.read()
    grec = cv2.cvtColor( rec , cv2.COLOR_BGR2GRAY)
    face_det = face_cascade.detectMultiScale(grec, scaleFactor=1.2 , minNeighbors=6 )
    for (x ,y ,w ,h ) in face_det :
        roi = rec[y:y+h , x:x+w]
        groi = grec[y:y+h , x:x+w]
        cv2.rectangle(rec , (x,y), (x+w,y+h) , (0,255,0) , 6)
        eye_det = eye_cascade.detectMultiScale(groi)
        for (xe , ye, we ,he ) in eye_det :
            cv2.rectangle(roi , (xe,ye),(xe+we, ye+he) , (255,0,0), 6)
        cv2.imshow('frame', rec)
        
    if cv2.waitKey(1) & 0xFF == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
