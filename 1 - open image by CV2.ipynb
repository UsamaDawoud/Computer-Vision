```python
ac_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(ac_img);
```

```python
# To read an image in gray scale 
img_gray = cv2.imread('IMG-20181229-WA0023.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray);
```

```python
# it is not white & black because of features of (imshow) and it's color maping 
# to show the image in white/black range ( adding ( cmap = 'gray'))
plt.imshow(img_gray, cmap = 'gray');
```

```python
img_gray.shape
```

```python
# It possible to resize the image by :
resized_image = cv2.resize(ac_img , (800 , 500))
plt.imshow(resized_image)
```

```python
# also can resize by using ratio 
# for example to resize it equal 65% from original image 
w_ratio = 0.65
h_ratio = 0.65
resized_ratio_image = cv2.resize(ac_img , (0,0), ac_img , w_ratio , h_ratio)
plt.imshow(resized_ratio_image );
```

```python
resized_ratio_image.shape
```

```python
# FLIP the images udside-down 
flipped1 = cv2.flip(ac_img , 0)
plt.imshow(flipped1);
```

```python
# if the flip = 1 (right to left  change)
flipped2 = cv2.flip(ac_img ,1)
plt.imshow(flipped2);
```

```python
# if the flip = -1  ( upside - down & right - left flip)
flipped3 = cv2.flip(ac_img ,-1)
plt.imshow(flipped3);
```

```python
# To save an image
cv2.imwrite('flipped.jpg',flipped3)
```

```python
flipped = cv2.imread('flipped.jpg')
plt.imshow(flipped);
```

```python
# Note the image 'flipped ' saved by cv2 !! (i.e. according to coding BGR )
# so to save it accoring to coding RGB
flipped2 = cv2.cvtColor(flipped , cv2.COLOR_BGR2RGB)
cv2.imwrite('flipped_2.jpg' , flipped2)
```

```python
flipped_2RGB = cv2.imread('flipped_2.jpg')
plt.imshow(flipped_2RGB)
```
