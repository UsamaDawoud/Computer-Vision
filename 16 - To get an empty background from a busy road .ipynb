# A code to get empty background of a busy road 


```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
video = cv2.VideoCapture('cars_road1.mp4')
FOI = video.get(cv2.CAP_PROP_FRAME_COUNT) * np.random.uniform(size =30) 

frames= []
for frameOI in FOI :
    video.set(cv2.CAP_PROP_POS_FRAMES , frameOI)
    _ , frame = video.read()
    
    frames.append(frame)
    
backgroundimage = np.median(frames , axis=0).astype(dtype=np.uint8)

backgroundimage= cv2.cvtColor(backgroundimage, cv2.COLOR_BGR2RGB)
plt.imshow( backgroundimage)

    
video.release()

    







```
