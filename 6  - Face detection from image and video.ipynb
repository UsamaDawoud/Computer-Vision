```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.cvtColor(cv2.imread('1.jpg'), cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img1, cmap='gray')
```

```python
img2 = cv2.cvtColor(cv2.imread('2.jpg'), cv2.COLOR_BGR2GRAY)
plt.imshow(img2 , cmap='gray')
```

```python
img3 = cv2.cvtColor(cv2.imread('3.jpg') , cv2.COLOR_BGR2GRAY)
plt.imshow(img3, cmap='gray')
```

```python
face_cascade= cv2.CascadeClassifier('C:/Users/MAN/a/deep learning/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_det(img):
    img_copy = img.copy()
    detected = face_cascade.detectMultiScale(img_copy)
    for (x,y,w,h) in detected:
        cv2.rectangle(img_copy , (x,y) ,(x+w , y+h) , (255,255,255) , 10)
    return img_copy    
```

```python
image1_det = face_det(img1)
plt.imshow(image1_det, cmap='gray')
```

```python
image2_det = face_det(img2)
plt.imshow(image2_det, cmap='gray')
```

```python
image3_det = face_det(img3)
plt.imshow(image3_det, cmap='gray')
```

```python
# in last image it can be seen that the program detect some shapes as faces !!! so to correct the performance, will use some aditional features 
def face_det_mod_1(img):
    img_copy = img.copy()
    detected = face_cascade.detectMultiScale(img_copy , scaleFactor=1.2 , minNeighbors=6)
    for (x,y,w,h) in detected:
        cv2.rectangle(img_copy , (x,y) ,(x+w , y+h) , (255,255,255) , 10)
    return img_copy   
```

```python
image2_det = face_det_mod_1(img2)
plt.imshow(image2_det, cmap='gray')
```

```python
image3_det = face_det_mod_1(img3)
plt.imshow(image3_det, cmap='gray')
```

```python
def face_det_mod_2(img):
    img_copy = img.copy()
    detected = face_cascade.detectMultiScale(img_copy , scaleFactor=1.3 , minNeighbors=6)
    for (x,y,w,h) in detected:
        cv2.rectangle(img_copy , (x,y) ,(x+w , y+h) , (255,255,255) , 10)
    return img_copy   
```

```python
image2_det = face_det_mod_2(img2)
plt.imshow(image2_det, cmap='gray')
```

```python
image3_det = face_det_mod_2(img3)
plt.imshow(image3_det, cmap='gray')
```

```python
# by increasing scaleFactor from 1.2 to 1.3 , in img2 it detect the only one face but in img3 it miss two faces which looking aside ! 
# so the algorithim is not perfect 
# in this seniro scaleFactor = 1.2 will be better (it better to detect non face as face than not detect available faces in the image )
# but the judgement still according to the uses ( for now i will cosider function face_det_mod_1 ) 
```

```python
# now will check the performance with video 
```

```python
cap = cv2.VideoCapture(0)

while True :
    ret , rec = cap.read()
    det_rec = face_det_mod_1(rec)
    cv2.imshow('frame' , det_rec)
    if cv2.waitKey(1) & 0xFF == 27 : 
        break
        
cap.release()
cv2.destroyAllWindows()
    
```
