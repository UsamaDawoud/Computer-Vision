```python
import cv2
```

```python
def setparameters (event, x,y,flags,parm): 
    global pt1,pt2,topclick,botclick


    if event == cv2.EVENT_LBUTTONDOWN :

        if botclick == True : 
            pt1 = (0,0)
            pt2 = (0,0)
            topclick = False
            botclick = False
            
        if topclick == False : 
            topclick = True
            pt1 = (x,y)
        
        else :
            botclick = True
            pt2 = (x,y)



# global Var : 

pt1 = (0,0)
pt2 = (0,0)
topclick = False
botclick = False



cap=cv2.VideoCapture(0)
cv2.namedWindow('test')
cv2.setMouseCallback('test', setparameters )


if cap.isOpened() == False:
    print(' No Video to show')
else:
    ret,frame= cap.read()
    while ret == True : 
            ret,frame= cap.read()
        
            
            if topclick == True and botclick == False : 
                cv2.circle(frame, pt1, color = (0,0,255),radius=3, thickness = -1)
            elif topclick == True and botclick == True : 
                cv2.rectangle(frame, pt1,pt2, color=(0,0,255), thickness=3)
        
            cv2.imshow('test',frame)
        
            if cv2.waitKey(1) != -1 :
                break     

cap.release()
cv2.destroyAllWindows()
```
