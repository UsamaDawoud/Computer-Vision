```python
# now calling the classifier that maded with small set of training images as experiment
```

```python
cascade_ff = cv2.CascadeClassifier('cascade_facefemale.xml')
```

```python
def detect_ff(image ):
    img = cv2.cvtColor(image  , cv2.COLOR_RGB2GRAY)
    a = cascade_ff.detectMultiScale(img, scaleFactor=1.2 , minNeighbors=6 )
    image_c = image.copy()
    for (x,y,w,h) in a :
        
        cv2.rectangle(image_c , (x,y) , (x+w , y+h) , (0,255,0) , 5)
    return image_c
```
